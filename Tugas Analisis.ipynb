# Tugas Analisis - Siswa Sekolah Menengah Atas yang Putus Sekolah tahun 2008-2011

Data di bawah ini adalah mengenai jumlah siswa SMA yang putus sekolah per Provinsi untuk periode tahun 2008 s.d. 2011.
Data yang ditampilkan adalah jumlah serta persentase siswa SMA yang putus sekolah.

sumber data: https://data.go.id/dataset/siswa-sekolah-menengah-atas-yang-putus-sekolah

```python
import pandas as pd

data = pd.read_csv('processed-data-putus-sekolah-sma-2008-2011.csv', delimiter = ',', encoding = "ISO-8859-1", dtype={'kode_hs':object})
```

```python
data
```

## Statistik data

### Jumlah Siswa SMA yang putus sekolah secara nasional per tahun

```python
data.groupby('tahun').sum()['jumlah_putus_sekolah']
```

### Siswa SMA yang putus sekolah per provinsi per tahun

```python
data.pivot(index='nama_provinsi', columns='tahun', values='jumlah_putus_sekolah')
```

### Rata-rata jumlah dan persentase siswa putus sekolah setiap tahun

```python
data.groupby('nama_provinsi')['jumlah_putus_sekolah','persen_putus_sekolah'].mean()
```

### Provinsi dengan rata-rata jumlah siswa putus sekolah terbanyak

```python
data.groupby('nama_provinsi')['jumlah_putus_sekolah'].mean().nlargest()
```

### Provinsi dengan rata-rata persentase siswa putus sekolah terbesar

```python
data.groupby('nama_provinsi')['persen_putus_sekolah'].mean().nlargest()
```

## Visualisasi

### Jumlah siswa putus sekolah secara nasional tahun 2008-2011

```python
ax = data.groupby('tahun').sum()['jumlah_putus_sekolah']
ax = ax.plot(stacked=True, title ="Jumlah Siswa Putus Sekolah Nasional", fontsize=12, figsize=(20,10))
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah", fontsize=12)
```

### Jumlah siswa putus sekolah per provinsi tahun 2008-2011

```python
ax2 = data.pivot(index='nama_provinsi', columns='tahun', values='jumlah_putus_sekolah')
ax2 = ax2.plot.bar(stacked=True, title ="Jumlah Siswa Putus Sekolah Per Provinsi", fontsize=12, figsize=(20,10))
ax2.set_xlabel("Tahun", fontsize=12)
ax2.set_ylabel("Jumlah", fontsize=12)
```

### Rata-rata persentase siswa putus sekolah tiap provinsi

```python
ax3 = data.groupby('nama_provinsi')['persen_putus_sekolah'].mean()
ax3 = ax3.plot.bar(stacked=True, title ="Rata-rata Persentase Siswa Putus Sekolah", fontsize=12, figsize=(20,10))
ax3.set_xlabel("Provinsi", fontsize=12)
ax3.set_ylabel("Persentase", fontsize=12)
```

## Narasi

Jumlah siswa SMA yang putus sekolah secara nasional cukup besar. Sebanyak 141712 siswa putus sekolah pada tahun 2008. Jumlah itu menurun pada tahun 2009 menjadi 126069 siswa pada 2009, namun kembali naik pada tahun 2010 menjadi 142275 siswa. Pada tahun 2011 jumlah siswa SMA putus sekolah menurun drastis menjadi 47709 siswa.

Yang perlu diperhatikan adalah sebgaian besar siswa SMA putus sekolah berasal dari provinsi-provinsi yang padat penduduk. Lima provinsi dengan jumlah siswa putus sekolah terbanyak adalah Jawa Timur, Jawa Barat, Sumatera Utara, Jawa Tengah, dan Sumatera Selatan. Sementara provinsi dengan persentase siswa putus sekolah terbesar adalah Nusa Tenggara Timur, Kalimantan Selatan, Kalimantan Barat, Kepulauan Bangka Belitung, dan Papua.

Pemerintah perlu memberikan fokus pada provinsi-provinsi tersebut, untuk mencari tahu penyebab tingginya tingkat siswa putus sekolah. Meskipun secara keseluruhan siswa yang putus sekolah menurun pada 2011.
